## Identifying Outliers with Skewness
Several machine learning algorithms make the assumption that the data follow a normal (or Gaussian) distribution. This is easy to check with the skewness value, which explains the extent to which the data is normally distributed. Ideally, the skewness value should be between -1 and +1, and any major deviation from this range indicates the presence of extreme values.

```python
df.skew()
```

## Outlier Treatment by using Quantile-based Flooring and Capping
In this technique, we will do the flooring (e.g., the 10th percentile) for the lower values and capping (e.g., the 90th percentile) for the higher values. The lines of code below print the 10th and 90th percentiles of the variable 'Clothing ID', respectively. These values will be used for quantile-based flooring and capping.


```python
print(df['Clothing ID'].quantile(0.10))
print(df['Clothing ID'].quantile(0.90))
```

```python
# Removing the outliers,then, we calculate the skewness value again, which comes out much better now.
df["Clothing ID"] = np.where(df["Clothing ID"] <820.0, 820.0,df['Clothing ID'])
df["Clothing ID"] = np.where(df["Clothing ID"] >1094.0, 1094.0,df['Clothing ID'])
print(df['Clothing ID'].skew())
```

```python
print(df['Rating'].quantile(0.10))
print(df['Rating'].quantile(0.90))
```

```python
df["Rating"] = np.where(df["Rating"] <2.0, 2.0,df['Rating'])
df["Rating"] = np.where(df["Rating"] >5.0, 5.0,df['Rating'])
print(df['Rating'].skew())
```

## Replacing Outliers with Median Values
In this technique, we replace the extreme values with median values. It is advised to not use mean values as they are affected by outliers.

```python
# Prints the 50th percentile value, or the median
print(df['Positive Feedback Count'].quantile(0.50)) 
# Prints the 95th percentile value
print(df['Positive Feedback Count'].quantile(0.95)) 
```

```python
# Replaces all those values in the 'Age' variable, which are greater than the 95th percentile, with the median value
df['Positive Feedback Count'] = np.where(df['Positive Feedback Count'] > 11.0, 1.0, df['Positive Feedback Count'])
df.describe()
```
